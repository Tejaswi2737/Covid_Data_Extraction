<h1> COVID-19 Data Extraction</h2>

Import the required modules and load the url

```python
import pandas as pd
url1='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
df = pd.read_csv(url1, error_bad_lines=False)
df
```

Data wrangling to obtain the required dataframe ie., the list of confirmed cases in all countries on all the dates

```python
df_conf=df.drop(columns=["Province/State","Lat","Long"])
df_conf=df_conf.groupby(df_conf["Country/Region"]).sum()
df_conf.reset_index(inplace=True)
df_conf
```

```python
df_conf=df_conf.transpose()
a=df_conf.iloc[0]
df_conf.columns=a
df_conf=df_conf.drop(index=["Country/Region"])
df_conf=df_conf.reset_index(inplace=False)
df_conf=df_conf.rename(columns={"index":"Date"}) 
countries=list(df_conf.columns)
countries=countries[1:len(countries)]
countries
```

Confirmed Cases Data and storing it into excel sheet

```python
df_conf.insert(1,"World",df_conf[countries].sum(axis=1))
df_conf.to_excel("Data_Confirmed.xls")
df_conf
```

Extracting the daily new cases confirmed from the total cases dataset and storing it in excel sheet

```python
df_new_conf=df_conf[countries].diff()
df_new_conf.iloc[0]=df_conf[countries].iloc[0]
df_new_conf.insert(0,"Date",df_conf["Date"])
df_new_conf.insert(1,"World",df_new_conf[countries].sum(axis=1))
df_new_conf.to_excel("Data_New_Confirmed.xls")
df_new_conf
```

Repeating the above process to obatin the total and daily death reported and storing them in an excel file 

```python
url2='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
df = pd.read_csv(url2, error_bad_lines=False)
df_death=df.drop(columns=["Province/State","Lat","Long"])
df_death=df_death.groupby(df_death["Country/Region"]).sum()
df_death.reset_index(inplace=True)
df_death=df_death.transpose()
a=df_death.iloc[0]
df_death.columns=a
df_death=df_death.drop(index=["Country/Region"])
df_death=df_death.reset_index(inplace=False)
df_death=df_death.rename(columns={"index":"Date"})
df_death.insert(1,"World",df_death[countries].sum(axis=1))
df_death.to_excel("Data_Deaths.xls")


df_new_deaths=df_death[countries].diff()
df_new_deaths.iloc[0]=df_death[countries].iloc[0]
df_new_deaths.insert(0,"Date",df_conf["Date"])
df_new_deaths.insert(1,"World",df_new_deaths[countries].sum(axis=1))
df_new_deaths.to_excel("Data_New_Deaths.xls")
```

Extracting the total and daily recoverries data and storing them in an excel file 

```python
url3='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
df = pd.read_csv(url3, error_bad_lines=False)
df_recov=df.drop(columns=["Province/State","Lat","Long"])
df_recov=df_recov.groupby(df_recov["Country/Region"]).sum()
df_recov.reset_index(inplace=True)
df_recov=df_recov.transpose()
a=df_recov.iloc[0]
df_recov.columns=a
df_recov=df_recov.drop(index=["Country/Region"])
df_recov=df_recov.reset_index(inplace=False)
df_recov=df_recov.rename(columns={"index":"Date"})
df_recov.insert(1,"World",df_recov[countries].sum(axis=1))
df_recov.to_excel("Data_Recovered.xls")

df_new_recov=df_recov[countries].diff()
df_new_recov.iloc[0]=df_recov[countries].iloc[0]
df_new_recov.insert(0,"Date",df_conf["Date"])
df_new_recov.insert(1,"World",df_new_recov[countries].sum(axis=1))
df_new_recov.to_excel("Data_New_Recover.xls")
```
